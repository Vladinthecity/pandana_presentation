```python
import pandas as pd
import pandana as pdna
%matplotlib inline
```

## Use OpenStreetMap networks for Bay Area that are included in the repo - someday soon we (GeoPandas) will have direct OSM import

```python
net_store = pd.HDFStore('osm_bayarea.h5', "r")
nodes = net_store.nodes
edges = net_store.edges
print len(nodes), "- count of nodes"
print nodes.head(3)
print len(edges), "- count of edges"
print edges.head(3)
```

## Initialize and preprocess the network - keep in mind distances are in meters

```python
net=pdna.Network(nodes.x, 
                       nodes.y, 
                       edges["from"], 
                       edges.to, 
                       edges[["weight"]])
net.precompute(3000)
```

## Read in the public San Francisco data

```python
store = pd.HDFStore('sanfran_public.h5')
store
```

## Parcels already have the node ids assigned.  Let's merge buildings to parcels and do an aggregation on buildings.

```python
buildings = pd.merge(store.buildings, store.parcels, left_on="parcel_id", right_index=True)
buildings.head(3)
```

## This column relates the buildings data to the nearest node_id in the network

```python
buildings._node_id.describe()
```

## Add the number of residential_units as an attribute of the network

```python
net.set(buildings._node_id, buildings.residential_units)
```

## This code sets a bounding box for the output visualization

```python
"""
from shapely.geometry import Point
from fiona.crs import from_epsg
import geopandas as gpd
bbox=[-122.539365,37.693047,-122.347698,37.816069]
bbox = gpd.GeoSeries([Point(bbox[0], bbox[1]),
                      Point(bbox[2], bbox[3])], 
                     crs=from_epsg(4326))
bbox = bbox.to_crs(epsg=3740)
bbox = [bbox[0].x, bbox[0].y, bbox[1].x, bbox[1].y]
"""

# the code above is to conver the lat-lng bounding box of San Francisco to srid 3740
# which is the srid of the network.  you can skip that step by just setting the
# bounding box in the right srid as I did below

bbox = [540610.6218789469, 4171858.652125045, 557413.4276436339, 4185608.2009922867]
```

## This is it - run the queries!

```python
%time res_units = net.aggregate(2000, type="sum", decay="linear")
```

## Describe the output numerically

```python
print res_units.describe()
```

## Hard to understand numerically, but easy to understand as a map.  Here's a map of the sum of residential_units with a linear decay.  This is displayed in matplotlib where green indicates larger numbers.

```python
net.plot(res_units, bbox=bbox, scheme="diverging", 
         color="BrBG", log_scale=False)
```

## Combining Pandas with Pandana is where much of the power of this approach comes from.  Here we query the buildings dataframe for buildings built more recently than 1980 and then do a network summation.  The "center of mass" has moved significantly as most constuction has occurred in SOMA and Mission Bay recently.

```python
modern_buildings = buildings.query('year_built > 1980')
net.set(modern_buildings._node_id, modern_buildings.residential_units)
%time modern = net.aggregate(2000, type="sum", decay="linear")
net.plot(modern, bbox=bbox, scheme="diverging", 
         color="BrBG", log_scale=False)
```

## Households can then be merged with buildings

```python
households = pd.merge(store.households, buildings, left_on="building_id", right_index=True)
households.head(3)
```

## And a similar query can be run to look for aggregations of high income households

```python
poor_households = households.query('income > 200000')
net.set(poor_households._node_id)
%time high_hhs = net.aggregate(2000, type="sum", decay="linear")
net.plot(high_hhs, bbox=bbox, scheme="diverging", 
         color="BrBG", log_scale=False)
```

## Create a nodes dataframe

```python
nodes = pd.DataFrame({
    'high_hhs': high_hhs,
    'res_units': res_units,
    'modern': modern
})
```

## Merge nodes and parcels and write the result to a csv for visualization in GeoCanvas

```python
pd.merge(store.parcels, nodes, left_on="_node_id", right_index=True, how="inner")\
    .to_csv("indicators.csv", index_label="parcel_id")
```
